```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
df = pd.read_csv('Hotel_bookings_final.csv')

df.head()
```

```python
print("Rows and Columns:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)
```

```python
df.isnull().sum()
```

```python
df.describe(include='all')
```

```python
df['booking_status'].value_counts()
```

```python
df['booking_status'].value_counts().plot(
    kind='bar',
    figsize=(8,5)
)

plt.title("Booking Status Distribution")
plt.xlabel("Status")
plt.ylabel("Count")
plt.show()
```

```python
channel_counts = df['booking_channel'].value_counts()

print(channel_counts)
```

```python
channel_counts.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title("Bookings by Channel")
plt.xlabel("Channel")
plt.ylabel("Bookings")
plt.show()
```

```python
room_counts = df['room_type'].value_counts()

print(room_counts)
```

```python
room_counts.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title("Bookings by Room Type")
plt.xlabel("Room Type")
plt.ylabel("Bookings")
plt.show()
```

```python
star_counts = df['star_rating'].value_counts()

print(star_counts)
```

```python
star_counts.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title("Bookings by Star Rating")
plt.xlabel("Star Rating")
plt.ylabel("Bookings")
plt.show()
```

```python
df['is_cancelled'] = np.where(
    df['booking_status']=='Cancelled',
    1,
    0
)
```

```python
cancel_rate = df['is_cancelled'].mean()*100

print("Cancellation Rate:", round(cancel_rate,2),"%")
```

```python
channel_cancel = pd.crosstab(
    df['booking_channel'],
    df['booking_status']
)

channel_cancel
```

```python
channel_cancel.plot(
    kind='bar',
    figsize=(10,6)
)

plt.title("Cancellation by Booking Channel")
plt.show()
```

```python
room_cancel = pd.crosstab(
    df['room_type'],
    df['booking_status']
)

room_cancel
```

```python
room_cancel.plot(
    kind='bar',
    figsize=(10,6)
)

plt.title("Cancellation by Room Type")
plt.show()
```

```python
df.groupby('booking_channel')['booking_value'].sum().sort_values(ascending=False)
```

```python
df.groupby('booking_channel')['booking_value'].sum().sort_values().plot(
    kind='barh',
    figsize=(8,5)
)

plt.title("Revenue by Booking Channel")
plt.show()
```

```python
df.groupby('room_type')['booking_value'].sum()
```

```python
df.groupby('room_type')['booking_value'].sum().plot(
    kind='bar',
    figsize=(8,5)
)

plt.title("Revenue by Room Type")
plt.show()
```

```python
df['booking_date'] = pd.to_datetime(df['booking_date'])

df['month'] = df['booking_date'].dt.month
```

```python
monthly = df.groupby('month')['booking_value'].sum()

monthly
```

```python
monthly.plot(
    marker='o',
    figsize=(10,5)
)

plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

```python
df['check_in_date'] = pd.to_datetime(df['check_in_date'])
df['check_out_date'] = pd.to_datetime(df['check_out_date'])

df['stay_length'] = (
    df['check_out_date'] - df['check_in_date']
).dt.days

df[['check_in_date','check_out_date','stay_length']].head()
```

```python
df['stay_length'].describe()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df['stay_length'],
    bins=20
)

plt.title("Stay Length Distribution")
plt.xlabel("Stay Length (Days)")
plt.ylabel("Number of Bookings")
plt.show()
```

```python
insights = {
    "Metric":[
        "Total Bookings",
        "Total Revenue",
        "Cancellation Rate"
    ],

    "Value":[
        len(df),
        round(df['booking_value'].sum(),2),
        round(df['is_cancelled'].mean()*100,2)
    ]
}

pd.DataFrame(insights)
```

```python
summary = pd.DataFrame(insights)

summary.to_csv(
    "Hotel_Booking_Summary.csv",
    index=False
)

print("Report Saved Successfully")
```
